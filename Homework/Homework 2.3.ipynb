### Homework 2.3 - 2D Temperature Distribution

Solve 2D heat equation for the following initial conditions:

$$\begin{array}{l}
u(x,0) = \sin (2\pi x)\\
u(x,1) = \sin (2\pi x)\\
u(0,y) = \sin (3\pi x)\\
u(1,y) = \sin (\pi x)
\end{array}$$

where $u(x, y)$ represent the steady state temperature distribution on 2D plane.
 
Plot the 2D heat map of the solution.

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.animation as animation
#%matplotlib inline
```

```python
N = 100
L = 1
x = np.linspace(0, L, num=N + 1, endpoint=True)
y = np.linspace(0, L, num=N + 1, endpoint=True)

Ux0 = np.sin(2*np.pi*x)
Ux1 = np.sin(2*np.pi*x)

U0y = np.sin(3*np.pi*y)
U1y = np.sin(np.pi*y)

U_ans = np.zeros((N+1,N+1))
U_ans[0,:] = U0y
U_ans[N,:] = U1y
U_ans[:,0] = Ux0
U_ans[:,N] = Ux1

max_it = 1000
diff = 1000

U_old = U_ans.copy()

while max_it > 0:
    if ((diff)/((N)**2)) <= 10**-5:
        print(max_it)
        break
    else:
        max_it -= 1
        for i in range(1,N):
            for j in range(1,N):
                U_ans[i, j] =  0.25*(U_ans[i+1, j] + U_ans[i-1, j] + U_ans[i, j+1] + U_ans[i, j-1])

        diff = np.sum(np.abs(U_ans - U_old))
        U_old = U_ans.copy()

plt.imshow(U_ans, cmap = "hot")
plt.show()
```
